```python
while True:
    cv2.imshow('result',img)
    #27 is ASCII of Esc
    if cv2.waitKey(2) == 27:
        break
cv2.destroyAllWindows()
```

```python
haar_data = cv2.CascadeClassifier('data.xml')
```

```python
haar_data.detectMultiScale(img)
#perform sliding window for haar features and will return no of faces in array
#first array depicting X,Y,Width and Height for first face
#second array depicting X,Y,Width and Height for second face and so on
```

```python
while True:
    faces = haar_data.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,255),4)
        #to draw rectangle of colour 255,0,255 i.e pink of border 4 on faces
    cv2.imshow('result',img)
    #27 is ASCII of Esc
    if cv2.waitKey(2) == 27:
        break
cv2.destroyAllWindows()
```

```python
capture = cv2.VideoCapture(0) #0 for default camera
data =[] #for storing face data
while True:
    flag,img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,255),4)
            #to draw rectangle of colour 255,0,255 i.e pink of border 4 on faces
            face = img[y:y+h,x:x+w, :]
            #for slicing of 3D array to get face image from entire img
            face = cv2.resize(face,(50,50))
            #for resizing face image into 50,50
            print(len(data))
            if len(data) < 200:
                data.append(face) 
                #to store 400 images in data array
             
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(data) >= 200:
            break
        
capture.release()        
cv2.destroyAllWindows()
```

```python
capture = cv2.VideoCapture(0)
#pass 0 for capturing images from default camera
#pass video.mp4 for capturing images from recorded video
data2 =[] #for storing face data
while True:
    flag,img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,255),4)
            #to draw rectangle of colour 255,0,255 i.e pink of border 4 on faces
            
            face = img[y:y+h,x:x+w, :]
            #for slicing of 3D array to get face image from entire img
            
            face = cv2.resize(face,(50,50))
            #for resizing face image into 50,50
            
            print(len(data2))
            if len(data2) < 200:
                data2.append(face) 
                #to store 200 images in data array
             
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(data2) >= 200:
            break
        
capture.release()        
cv2.destroyAllWindows()
```

```python
np.save('with_mask.npy',data2)
```
